SMT (Statistical Machine Translation) — это подход к машинному переводу, основанный на статистических моделях. Цель — найти наиболее вероятный перевод фразы на языке Y, исходя из фразы на языке X.

Задача: Найти наиболее вероятный перевод текста на языке X на язык Y через максимизацию вероятности P(y|x), т.е. argmax P(y|x), где y — перевод на языке Y, а x — исходное предложение на языке X.
Ключевые этапы:
Моделирование вероятности предложений в языке Y
P(y) с помощью языковой модели.

Компоненты:
- Моделирование вероятности перевода с помощью модели перевода на уровне слов или фраз в языке Y
- Поиск переводческих соответствий X - Y и моделирование их с помощью теории вероятности (наиболее правдоподобные соответствия)
- Декодинг для поиска наиболее вероятного перевода

Параллельные корпуса
Выравнивание: сопоставление фраз или слов в параллельных корпусах для нахождения соответствий между языками (статистика)
Языковая модель (LM)
Модель перевода
Декодер

4. Моделирование SMT
- Модель IBM 1, использующая алгоритм Expectation-Maximization (EM) для подсчета вероятности того, что случайное слово x_vocab соответсвует случайному слову y_vocab
- t-model для итерации по обучающей выборке: каждое слово из исходного языка X получает соответствующее слово на языке Y с определенной вероятностью в виде таблицы
- Итерации по обучающей выборке для улучшения точности моделей
- Обучение: создание слотов для подсчета статистики, Expactation (предварительная статистика), Maximization
- Биграммная языковая модель (LM): предсказывание вероятности появления следующего слова в зависимости от предыдущего.
